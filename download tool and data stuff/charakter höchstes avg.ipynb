```python
import seaborn as sns
import matplotlib.pyplot as plt 
import statistics
import numpy as np
import re
import pandas as pd
```

```python

df_alt = pd.read_csv('sortiert.csv')

top_50 = (list(df_alt['charakter'].str.split(',').explode().value_counts().nlargest(25).items()))

chrakter,anzahl = np.array(top_50).T.tolist()
anzahl = list(map(int, anzahl))
df = pd.DataFrame(list(zip(chrakter,anzahl)), columns=['Charakter', 'Anzahl'])
```

```python
i = 0
median = 0
index_median_score = []
score_character = []
for c in chrakter:
    for index,row in df_alt.iterrows():
        if c in str(row["charakter"]):
                index_median_score.append(row["score"])
    print(statistics.median(index_median_score))
    score_character.append(statistics.median(index_median_score))
    print(c)
    index_median_score = []
    i+= 1
    median = 0
```

```python
df = pd.DataFrame(list(zip(chrakter,anzahl, score_character)), columns=['Charakter', 'Anzahl', 'Likes'])
df = df.sort_values('Likes', ascending=False)
digagramm = sns.barplot(x='Charakter', y='Likes',data=df).set_title("R34 Top 25 most common Star Wars characters by avg likes")
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='small'
)
#plt.ylim(0, 10)
#diagramm_fig = digagramm.get_figure()
#diagramm_fig.savefig('Chakrater_top25score.png')

#plt.show()
```
