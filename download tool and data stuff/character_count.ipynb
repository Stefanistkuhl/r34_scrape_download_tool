```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python

df = pd.read_csv('sorted.csv')

top_50 = (list(df['charakter'].str.split(',').explode().value_counts().nlargest(25).items()))

chrakter,anzahl = np.array(top_50).T.tolist()
anzahl = list(map(int, anzahl))
df = pd.DataFrame(list(zip(chrakter,anzahl)), columns=['Charakter', 'Anzahl'])
```

```python
df = pd.DataFrame(list(zip(chrakter,anzahl)), columns=['Charakter', 'Anzahl'])

digagramm = sns.barplot(x='Charakter', y='Anzahl',data=df).set_title("R34 Top 25 Star Wars Charkatere nach Anzahl")
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='small'
)
#plt.ylim(0, 10)
diagramm_fig = digagramm.get_figure()
#diagramm_fig.savefig('Chakrater_top25.png')

#plt.show()
```
